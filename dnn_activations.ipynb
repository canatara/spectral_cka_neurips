## MajajHong


```python
regions = ['V4', 'IT']
subject_names = ['Tito', 'Chabo']

# Get a matrix of responses for each trial
data_majajhong = {}
df_img_majajhong = {}
for region in regions:
    for subject in subject_names:

        df, df_img = data_utils.get_brainscore(region, subject_name=subject, avg_trials=True, get_stimuli=False)
        data_majajhong[(region, subject)] = df
        df_img_majajhong[(region, subject)] = df_img
        print((region, subject), df.shape, df_img.shape)

for key, value in df_img_majajhong.items():
    print(key, (value == df_img).all())
```

```python
dataset_name = 'majajhong'
model_name = 'resnet18'
pretrained = True

act = get_activations_from_df_img(df_img, model_name=model_name, pretrained=pretrained,
                                  dataset_name=dataset_name, device=device, batch_size=32,
                                  activations_root=ACTIVATIONS_ROOT)

for key, value in act.items():
    print(key, value.shape)
```
